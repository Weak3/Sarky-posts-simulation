```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
post_df=pd.read_csv("post.csv")
```

```python
plt.axis(ymax=0.06,xmin=0.7,xmax=1)

for f in range(0,9):
    plt.plot(post_df.P_Prob[post_df.Fails_reported==f],post_df.density[post_df.Fails_reported==f])
plt.figlegend(range(0,9))
plt.xlabel("P_pass")
```

```python
rep_errs_df=pd.read_csv("rep_errs.csv")
```

```python
plt.bar(rep_errs_df['Reported errors'],rep_errs_df[' Est. Prob.'])
plt.xlabel('EFP')
plt.show()
```

```python
mean_prob_df=pd.read_csv("mean_prob.csv")
plt.ylim(0.7,1)
plt.axline((0,0.85),(8,0.85),ls=":")
plt.bar(mean_prob_df.iloc[:,0],mean_prob_df.iloc[:,1])
plt.xlabel('EFP')
plt.ylabel('Exp. P_pass')
plt.show()
```
